```python
#Fuzzification penghasilan
def tinggi(penghasilan): 
    if penghasilan >= 11 :
        return 1
    elif penghasilan <= 8 :
        return 0
    elif penghasilan < 11 and penghasilan > 8 :
        return 2

def biasa(penghasilan):
    if penghasilan <= 9 and penghasilan >= 7 :
        return 1
    elif penghasilan >= 10 or penghasilan <= 5  :
        return 0
    elif (penghasilan < 10 and penghasilan > 9) or (penghasilan < 7 and penghasilan > 5) :
        return 2
    
def rendah(penghasilan):
    if penghasilan <= 5 and penghasilan >= 3:
        return 1
    elif penghasilan >= 6 or penghasilan <= 2:
        return 0
    elif (penghasilan < 6 and penghasilan > 5) or (penghasilan < 3 and penghasilan > 2) :
        return 2

def sangatrendah(penghasilan) :
    if penghasilan <= 2 :
        return 1
    elif penghasilan >= 4 :
        return 0
    elif penghasilan >= 3 and penghasilan < 4:
        return 2
```

```python
#fuzzification pengeluaran
def banyak(pengeluaran):
    if pengeluaran >= 9 :
        return 1
    elif pengeluaran <= 5 :
        return 0
    elif pengeluaran < 9 and pengeluaran > 5 :
        return 2

def sedang(pengeluaran):
    if pengeluaran <= 7 and pengeluaran >= 4:
        return 1
    elif pengeluaran >= 10 or pengeluaran <= 2 :
        return 0
    elif (pengeluaran > 7 and pengeluaran < 10) or (pengeluaran > 2 and pengeluaran < 4):
        return 2

def sedikit(pengeluaran):
    if pengeluaran <= 2 :
        return 1
    elif pengeluaran >= 3 :
        return 0
    elif pengeluaran > 2 and pengeluaran < 3 :
        return 2
```

```python
#nanti status dari setiap irisan akan ditampung pada array misal
#tinggi sedang rendah sangat rendah
#[2,2,0,0]
#berarti penghasilan diantara irisan biasa dan tinggi
def fuzzificationPenghasilan(gaji):
    statPenghasilan = [[],[]]
    if tinggi(gaji) != 0 :
        if tinggi(gaji) == 1:
            statPenghasilan[0].append("tinggi")
            statPenghasilan[1].append(1)
        elif tinggi(gaji) == 2:
            statPenghasilan[0].append("tinggi")
            fuzzification = -1*((gaji-11)/(11-8))
            statPenghasilan[1].append(fuzzification)

    if biasa(gaji)!= 0:
        if biasa(gaji) == 1:
            statPenghasilan[0].append("biasa")
            statPenghasilan[1].append(1)
        elif biasa(gaji) == 2:
            statPenghasilan[0].append("biasa")
            a = 5  #batas bawah 1
            b = 7 #batas atas 1
            c = 9 #batas atas 2
            d = 10 #batas bawah 2
            if gaji < b and gaji > a :
                fuzzification = ((gaji-a)/(b-a))
            elif gaji < d and gaji > c :
                fuzzification = -1*((gaji-d)/(d-c))

            statPenghasilan[1].append(fuzzification)

    if rendah(gaji) != 0:
        if rendah(gaji) == 1:
            statPenghasilan[0].append("rendah")
            statPenghasilan[1].append(1)
            
        elif rendah(gaji) == 2:
            statPenghasilan[0].append("rendah")
            a = 2 #batas bawah 1
            b = 3 #batas atas 1
            c = 5 #batas atas 2
            d = 6 #batas bawah 2
            if gaji < b and gaji > a :
                fuzzification = ((gaji-a)/(b-a))
            elif gaji < d and gaji > c:
                fuzzification = -1*((gaji-d)/(d-c))
            
            statPenghasilan[1].append(fuzzification)

    if sangatrendah(gaji) !=0:
        if sangatrendah(gaji) == 1:
            statPenghasilan[0].append("sangatrendah")
            statPenghasilan[1].append(1)

        elif sangatrendah(gaji) == 2:
            statPenghasilan[0].append("sangatrendah")
            fuzzification = -1*(gaji-3)/(3-2)
        
            statPenghasilan[1].append(fuzzification)

    return statPenghasilan


def fuzzificationPengeluaran(pengeluaran):
    statPengeluaran = [[],[]]
    if banyak(pengeluaran) != 0 :
        if banyak(pengeluaran) == 1:
            statPengeluaran[0].append("banyak")
            statPengeluaran[1].append(1)
        elif banyak(pengeluaran) == 2:
            statPengeluaran[0].append("banyak")
            fuzzification = -1*((pengeluaran-9)/(9-5))
            statPengeluaran[1].append(fuzzification)

    if sedang(pengeluaran)!= 0:
        if sedang(pengeluaran) == 1:
            statPengeluaran[0].append("sedang")
            statPengeluaran[1].append(1)
        elif sedang(pengeluaran) == 2:
            statPengeluaran[0].append("sedang")
            a = 2 #batas bawah 1
            b = 4 #batas atas 1
            c = 7 #batas atas 2
            d = 10 #batas bawah 2
            if pengeluaran > a and pengeluaran < b :
                fuzzification = ((pengeluaran-a)/(b-a))
            elif pengeluaran > c and pengeluaran < d:
                fuzzification = -1*((pengeluaran-d)/(d-c))
            statPengeluaran[1].append(fuzzification)

    if sedikit(pengeluaran) != 0 :
        if sedikit(pengeluaran) == 1:
            statPengeluaran[0].append("sedikit")
            statPengeluaran[1].append(1)
        elif sedikit(pengeluaran) == 2:
            statPengeluaran[0].append("sedikit")
            fuzzification = ((pengeluaran-2)/(3-2))
            statPengeluaran[1].append(fuzzification)

    return statPengeluaran
    
#penghasilan = fuzzificationPenghasilan(7)
#print(penghasilan)
#print(fuzzificationPengeluaran(5.7))
```

```python
#inference
def inference(penghasilan_sangatrendah,penghasilan_rendah,penghasilan_biasa,penghasilan_tinggi,pengeluaran_sedikit,pengeluaran_sedang,pengeluaran_banyak):
    
    rules = [[min(penghasilan_tinggi,pengeluaran_banyak),'layak'],[min(penghasilan_tinggi,pengeluaran_sedang),'tidak'],[min(penghasilan_tinggi,pengeluaran_sedikit),'tidak'],[min(penghasilan_biasa,pengeluaran_banyak),'layak'],[min(penghasilan_biasa,pengeluaran_sedang),'layak'],[min(penghasilan_biasa,pengeluaran_sedikit),'tidak'],[min(penghasilan_rendah,pengeluaran_banyak),'layak'],[min(penghasilan_rendah,pengeluaran_sedang),'layak'],[min(penghasilan_rendah,pengeluaran_sedikit),'tidak'], [min(penghasilan_sangatrendah,pengeluaran_banyak),'layak'],[min(penghasilan_sangatrendah,pengeluaran_sedang),'layak'],[min(penghasilan_sangatrendah,pengeluaran_sedikit),'layak']]

    listlayak=[]
    listtidaklayak=[]
    for i in range(len(rules)):
        if rules[i][1] == 'layak':
            listlayak.append(rules[i])
        elif rules[i][1] == 'tidak':
            listtidaklayak.append(rules[i])
    
    #disjunction
    m1=0
    m2=0
    for i in range(len(listlayak)):
        if m1 < listlayak[i][0]:
            m1 = listlayak[i][0]

    for i in range(len(listtidaklayak)):
        if m2 < listtidaklayak[i][0]:
            m2 = listtidaklayak[i][0]

    hasil = [m1,m2]
    return hasil
            
```

```python
#defuzzyfication
def defuzzyfication(layak,tidaklayak):
    if layak+tidaklayak != 0:
        return (((layak*100)+(tidaklayak*40))/(layak+tidaklayak))
    #else:
       # return 4
```

```python
def ProgramFuzzyLogic(a,b) :
    penghasilan = fuzzificationPenghasilan(a)
    pengeluaran = fuzzificationPengeluaran(b)
    penghasilan_tinggi = 0
    penghasilan_biasa = 0
    penghasilan_rendah = 0
    penghasilan_sangatrendah = 0
    pengeluaran_banyak = 0
    pengeluaran_sedang = 0
    pengeluaran_sedikit = 0

    for i in range(len(penghasilan[0])):
        if penghasilan[0][i] == "tinggi":
            penghasilan_tinggi = penghasilan[1][i]
        if penghasilan[0][i] == "biasa":
            penghasilan_biasa = penghasilan[1][i]
        if penghasilan[0][i] == "rendah":
            penghasilan_rendah = penghasilan[1][i]
        if penghasilan[0][i] == "sangat rendah":
            penghasilan_sangatrendah = penghasilan[1][i]

    for i in range(len(pengeluaran[0])):
        if pengeluaran[0][i] == "banyak":
            pengeluaran_banyak = pengeluaran[1][i]
        if pengeluaran[0][i] == "sedang":
            pengeluaran_sedang = pengeluaran[1][i]
        if pengeluaran[0][i] == "sedikit":
            pengeluaran_sedikit = pengeluaran[1][i]

    x = inference(penghasilan_sangatrendah,penghasilan_rendah,penghasilan_biasa,penghasilan_tinggi,pengeluaran_sedikit,pengeluaran_sedang,pengeluaran_banyak)

    defuzzy = defuzzyfication(x[0],x[1])
    #print(x[0])
    #print(x[1])
    return defuzzy
    
```
